## 2. Funciones: Creacion de Herramientas a Medida

Ya sabemos crear funciones, pero ahora vamos a usarlas como lo hacen los profesionales: para encapsular una lÃģgica especÃ­fica y hacer nuestro cÃģdigo mÃĄs limpio y reutilizable. Una buena funciÃģn hace **una sola cosa y la hace bien**.

Vamos a crear una funciÃģn que limpie un texto y otra que, usando la primera, cuente las palabras.

```python
# FunciÃģn 1: Especializada en limpiar texto
def limpiar_texto(texto):
    """Limpia un texto removiendo puntuaciÃģn comÃšn, incluidas marcas del espaÃąol."""
    # Nota: en espaÃąol necesitamos incluir Âŋ y ÂĄ ademÃĄs de los signos estÃĄndar.
    # TambiÃĐn eliminamos comillas y parÃĐntesis para obtener palabras limpias.
    caracteres_a_eliminar = ['.', ',', '!', '?', ';', ':', 'Âŋ', 'ÂĄ', '"', "'", '(', ')']
    texto_limpio = texto
    for char in caracteres_a_eliminar:
        texto_limpio = texto_limpio.replace(char, '')
    return texto_limpio.lower()  # Devuelve el texto limpio y en minÃšsculas
```

```python
# FunciÃģn 2: Especializada en contar palabras
def contar_palabras(texto):
    """Usa limpiar_texto() para procesar el texto y luego cuenta las palabras."""
    texto_procesado = limpiar_texto(texto) # ÂĄReutilizamos nuestra primera funciÃģn!
    palabras = texto_procesado.split()
    return len(palabras)
```

```python
# --- Uso de las funciones ---
mi_texto = "ÂĄHola, mundo! ÂŋListo/a para aprender sobre Procesamiento del Lenguaje Natural?"
cantidad = contar_palabras(mi_texto)

print(f"El texto original era: '{mi_texto}'")
print(f"DespuÃĐs de limpiarlo y contarlo, tiene {cantidad} palabras.")
```

## 3. ProgramaciÃģn Orientada a Objetos (POO): Un CÃģdigo mÃĄs Inteligente

La POO es un paradigma para organizar el cÃģdigo. La idea es simple: en lugar de tener datos y funciones por separado, los agrupamos en "objetos" que tienen tanto los datos (atributos) como las funciones que operan sobre esos datos (mÃĐtodos).

* **Clase**: Es el "plano" o la "receta". Define cÃģmo serÃĄn los objetos.
* **Objeto (o instancia)**: Es la "casa" construida a partir del plano. Es una entidad real con la que interactuamos.
* **`__init__(self, ...)`**: El "constructor". Es un mÃĐtodo especial que se ejecuta al crear un nuevo objeto para inicializar sus atributos.
* **`self`**: Se refiere al objeto mismo. AsÃ­ es como el objeto accede a sus propios datos y mÃĐtodos.

Vamos a crear una clase `ProcesadorTexto` que encapsule un texto y todas las operaciones que queremos hacer con ÃĐl.

```python
# DefiniciÃģn ÃNICA de la clase. Esta celda es la referencia para todo el notebook.
class ProcesadorTexto:
    """Encapsula un texto y expone operaciones de anÃĄlisis lingÃžÃ­stico bÃĄsico."""

    def __init__(self, texto):
        # Nota sobre limpieza: strip() solo elimina caracteres en los extremos del string.
        # Para limpiar puntuaciÃģn en todo el texto usamos replace(), igual que en limpiar_texto().
        print(f"Creando procesador para: '{texto[:40]}...'")
        self.texto_original = texto
        caracteres_a_eliminar = ['.', ',', '!', '?', ';', ':', 'Âŋ', 'ÂĄ', '"', "'", '(', ')']
        texto_limpio = texto.lower()
        for char in caracteres_a_eliminar:
            texto_limpio = texto_limpio.replace(char, '')
        self.texto_limpio = texto_limpio
        self.palabras = self.texto_limpio.split()

    def contar_palabras(self):
        """Devuelve el total de palabras (con repeticiones)."""
        return len(self.palabras)

    def contar_palabras_unicas(self):
        """Devuelve la cantidad de palabras distintas. `set` elimina duplicados."""
        return len(set(self.palabras))

    def calcular_frecuencia(self):
        """Devuelve un diccionario con la frecuencia de cada palabra."""
        frecuencia = {}
        for palabra in self.palabras:
            frecuencia[palabra] = frecuencia.get(palabra, 0) + 1
        return frecuencia
```

```python
# --- Uso de la clase ---
# Creamos un objeto (una instancia) de nuestra clase
procesador = ProcesadorTexto("Python para NLP es genial porque Python es versÃĄtil.")

# Ahora podemos usar sus mÃĐtodos
print(f"Total de palabras: {procesador.contar_palabras()}")
print(f"Palabras Ãšnicas: {procesador.contar_palabras_unicas()}")
print(f"Frecuencia de palabras: {procesador.calcular_frecuencia()}")
```
