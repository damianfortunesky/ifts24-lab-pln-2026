```python
import spacy
from spacy import displacy
```

```python
nlp = spacy.load("es_core_news_sm")
```

## Caso 1: tokenización y análisis básico

```python
doc = nlp(u'Yo, Matías Barreto, estoy volando hacia Buenos Aires.')
print([w.text for w in doc])
```

```python
# Cantidad de tokens en la oración
print("La cantidad de tokens es: ", len(doc))
```

```python
# Mostrar los tokens del documento
print("Los tokens son:\n" + "_" * 20 + "\n")
for token in doc:
    print(token)
```

```python
for token in doc:
    print(token.text, token.lemma_)
```

```python
# Iteramos sobre los tokens del documento y mostramos su texto, su etiqueta POS y la explicación correspondiente
for token in doc:
    print(token.text, token.pos_, spacy.explain(token.pos_))
```

```python
from collections import Counter
```

```python
word_freq = Counter(doc)
print(word_freq)
```

## Caso 2: lemas y dependencias

```python
doc = nlp(u'En abril de 2025 estaremos realizando una introducción a spaCy para trabajar conceptos básicos de PLN')
for token in doc:
    print(token.text, token.lemma_)
```

```python
from spacy.lang.es.examples import sentences

doc = nlp(sentences[11])
print(doc.text)
for token in doc:
    print(token.text, token.pos_, token.dep_)
```

```python
word_freq = Counter(doc)
print(word_freq)
```

```python
displacy.render(doc, style='dep', jupyter=True, options={'distance': 120})
```

## Caso 5: personalizar la lematización

```python
doc = nlp(u"Estoy volando hacia Baires")
print([w.text for w in doc])
```

```python
from spacy.language import Language
```

```python
# Componente personalizado para asignar lemas específicos
@Language.component("custom_lemmatizer")
def custom_lemmatizer(doc):
    for token in doc:
        if token.text == "Baires":
            token.lemma_ = "Buenos Aires"
    return doc

# Si ya existe, quitamos el componente personalizado
if "custom_lemmatizer" in nlp.pipe_names:
    nlp.remove_pipe("custom_lemmatizer")

# Agregamos el componente al pipeline
nlp.add_pipe("custom_lemmatizer", after="lemmatizer")

# Probamos el resultado con un ejemplo en español
doc = nlp(u"Estoy volando a Baires")
print([w.text for w in doc])
print([w.lemma_ for w in doc])
```
