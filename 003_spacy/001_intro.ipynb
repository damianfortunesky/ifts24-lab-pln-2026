## Parte 1: el problema - para la computadora, el texto es solo una "sopa de letras"

```python
texto = "La CEO de ACME, Lisa Chang, viajará a México para supervisar los nuevos proyectos de energía renovable. Ella se mostró muy entusiasmada."
print(texto)
```

¿Cómo contamos las palabras en este texto?

`len(texto)` nos da la cantidad de caracteres, así que no alcanza. Podés probar con `texto.split()`. Veamos qué pasa.

```python
palabras = texto.split()
print(palabras)
```

### Limitaciones: por qué necesitamos algo mejor

- `"ACME,"` y `"Chang,"` conservan la coma. No son palabras limpias.
- `"La"` y `"los"` son artículos, mientras que `"viajará"` es una forma conjugada del verbo `"viajar"`. ¿Cómo llevás cada palabra a su forma base para analizar frecuencias o temas?
- ¿Cómo identifica la máquina que `"ACME"` es una organización y que `"Lisa Chang"` es una persona? Para Python, por ahora, todo sigue siendo texto plano.

## Parte 2: la solución - spaCy como intérprete lingüístico

```python
!pip install spacy -q
!python -m spacy download es_core_news_sm -q
```

```python
import spacy
```

```python
# Cargamos el modelo de español
nlp = spacy.load("es_core_news_sm")

# Procesamos el texto
doc = nlp(texto)

print(doc)
```

## Parte 3: desglosar el objeto `Doc` - fundamentos de PLN en acción

**Tokenización: la unidad mínima de análisis**

Definición: un token es la unidad mínima de texto con significado, como una palabra, un número o un signo de puntuación.

```python
for token in doc:
    print(token.text)
```

**Lematización: la forma base de cada palabra**

Definición: el lema es la forma canónica o de diccionario de una palabra. Permite agrupar variantes que comparten significado.

```python
print("PALABRA\t\tLEMA")
print("-------\t\t----")
for token in doc:
    # Imprimimos el texto del token y su lema
    print(f"{token.text:<15}\t{token.lemma_}")
```

**Etiquetado gramatical (POS): el rol de cada palabra**

Definición: consiste en asignar a cada palabra su función gramatical, por ejemplo sustantivo, verbo o adjetivo.

```python
print("PALABRA\t\tPOS\t\tEXPLICACIÓN")
print("-------\t\t---\t\t-----------")
for token in doc:
    print(f"{token.text:<15}\t{token.pos_:<10}\t{spacy.explain(token.pos_)}")
```
