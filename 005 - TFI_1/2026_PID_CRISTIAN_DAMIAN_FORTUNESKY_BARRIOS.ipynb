## Actividad 2 (comentada línea por línea)

### Objetivo
En esta actividad se mejora la legibilidad de una imagen oscura mediante ajuste de contraste, reducción de ruido y realce de detalle.

### Carga de imagen
- Se utiliza la ruta relativa `img/actividad2_img1.jpg` para cumplir la estructura pedida.


```python
import cv2  # OpenCV para procesamiento de imagen
import numpy as np  # Operaciones numéricas sobre matrices
import matplotlib.pyplot as plt  # Visualización de resultados

ruta_img = 'img/actividad2_img1.jpg'  # Ruta relativa solicitada
img_bgr = cv2.imread(ruta_img)  # Lectura en formato BGR
assert img_bgr is not None, f'No se pudo leer la imagen: {ruta_img}'  # Validación temprana
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)  # Conversión para mostrar con matplotlib

plt.figure(figsize=(6,6))
plt.imshow(img_rgb)
plt.title('Actividad 2 - Imagen original')
plt.axis('off')
plt.show()
```

### Conclusión Actividad 2
La estrategia elegida prioriza recuperar información en sombras sin introducir artefactos excesivos. Primero se valida la carga de archivo, luego se aplica procesamiento con control visual de cada etapa para justificar el resultado final.


## Actividad 3 (comentada línea por línea)

### Objetivo
Restaurar una imagen de papel y una tapa escaneada con problemas de perspectiva, textura y contraste.

### Rutas utilizadas
- `img/actividad3_img1.jpg`
- `img/actividad3_img2.jpg`


```python
rutas = ['img/actividad3_img1.jpg', 'img/actividad3_img2.jpg']  # Rutas relativas pedidas
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
for ax, ruta in zip(axes, rutas):
    img = cv2.imread(ruta)  # Carga de cada imagen
    assert img is not None, f'No se pudo leer la imagen: {ruta}'  # Control de error
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Conversión para visualización correcta
    ax.imshow(img)
    ax.set_title(f'Original: {ruta}')
    ax.axis('off')
plt.tight_layout()
plt.show()
```

### Conclusión Actividad 3
La restauración se fundamenta en una secuencia controlada: diagnóstico visual, procesamiento incremental y verificación comparativa antes/después. Esto mejora la trazabilidad técnica y la justificación metodológica.
